```python
#2. Analisi delle Caratteristiche Generali
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('udemy_online_education_courses_dataset.csv')

total_courses = len(df)
total_subscribers = df['num_subscribers'].sum()
total_reviews = df['num_reviews'].sum()
unique_subjects = df['subject'].nunique()
price_range = (df['price'].min(), df['price'].max())
published_years = (pd.to_datetime(df['published_timestamp']).dt.year.min(),
                   pd.to_datetime(df['published_timestamp']).dt.year.max())

stats = {
    'Total Courses': f"{total_courses:,}",
    'Total Subscribers': f"{total_subscribers:,}",
    'Total Reviews': f"{total_reviews:,}",
    'Number of Subjects': unique_subjects,
    'Price Range': f"${price_range[0]} - ${price_range[1]}",
    'Publication Years': f"{published_years[0]} - {published_years[1]}"
}

plt.figure(figsize=(10, 5))

plt.text(0.1, 0.9, "Udemy Dataset Overview", fontsize=20, fontweight='bold')

vertical_spacing = 0.8
for key, value in stats.items():
    plt.text(0.1, vertical_spacing, f"{key}: {value}", fontsize=14)
    vertical_spacing -= 0.1

plt.axis('off')

plt.tight_layout()
plt.savefig('Overview_Text.png')
plt.show()
```

```python
subject_counts = df['subject'].value_counts()

plt.figure(figsize=(10, 5))
plt.pie(
    subject_counts, 
    labels=subject_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('colorblind', n_colors=len(subject_counts)),
    textprops={'fontsize': 12},
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}
)

plt.title('Subjects Distribution', fontsize=16, fontweight='bold', loc='center')

plt.tight_layout()
plt.savefig('subjects_pie_chart.png', dpi=300, transparent=True)
plt.show()
```
